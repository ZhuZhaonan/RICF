```python
import pandas as pd
import requests
import unicodedata
exec(requests.get('https://raw.githubusercontent.com/ma-ji/RICF/master/ricf_oid/ricf_lookup.py').text)
```

### 2013_basic_profile.

```python
df_13_basic=pd.read_csv('../../2013/2013_basic_profile.tsv', sep='\t', header=1).rename(columns={'ricf_oid':'ricf_oid_2013'})
df_13_basic.sample(3)
```

```python
for index in df_13_basic.index:
    ba_cn=unicodedata.normalize('NFKC', df_13_basic.loc[index, 'ba_cn'])
    ricf_oid=ricf_lookup(foundation_name=ba_cn)['ricf_oid']
    nacao_uscc=ricf_lookup(foundation_name=ba_cn)['nacao_uscc']
    df_13_basic.loc[index, 'ricf_oid']=ricf_oid
    df_13_basic.loc[index, 'ba_uscc']=nacao_uscc
```

```python
df_13_basic.sample(3)
```

```python
df_13_basic.to_csv('../../2013/2013_basic_profile.tsv', sep='\t', index=False, encoding='utf8')
```
