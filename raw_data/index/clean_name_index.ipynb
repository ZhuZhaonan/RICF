```python
import pandas as pd
```

```python
df_fund_catalog=pd.read_csv('foundation_catalog_171219.tsv', sep='\t', index_col=0)
```

```python
df_fund_catalog
```

```python
len(df_fund_catalog['inst_nm'].tolist()), len((df_fund_catalog['inst_nm'].tolist()))
```

```python
df_name_index=pd.read_csv('name_index.tsv', sep='\t')
```

```python
error_index=[]
for index in df_name_index.index:
    try:
        nm=df_name_index.loc[index, 'ba_cn']
        nm_offi=df_fund_catalog[df_fund_catalog.inst_nm==nm]['inst_nm'].values[0]
        uscc=df_fund_catalog[df_fund_catalog.inst_nm==nm]['uscc'].values[0]
        df_name_index.loc[index, 'ba_uscc']=uscc
        df_name_index.loc[index, 'nm_offi']=nm_offi
    except:
        error_index.append(index)
```

```python
len(error_index)
```
