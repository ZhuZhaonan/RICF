## Database 1 : RICF

RICF is the research infrastructure of Chinese foundations, a database for Chinese civil society studies. The data are crawled, parsed, and compiled manually or automatically by computer programs (Python Scrapy and other data processing packages, e.g., Pandas) from the following six sources, which are ranked by their credibility:

*  Annual reports and audited financial reports . 
* Information disclosed by supervising government departments .
* Information disclosed by the China Foundation Database (http://chinafoundation.org.cn; an information-disclosing platform supervised by the Civil Organization Administration Bureau, closed in early 2016 for unknown reasons).
* Information disclosed by the China Foundation Center (http://foundationcenter.org.cn; an  information-disclosing platform run by a nonprofit organization) .
*  News from the foundation’s official website. The website snapshots are taken and stored under the “raw data” folder (see Data Records section below; the same for source #6).
* News from credible magazines or websites 

The source , codebook and data format of this database.

 * Codebooks: https://github.com/ma-ji/RICF/blob/master/RICF%20Codebook.xlsx
 *  Data formats：string


##  Database 2: IRS 990 Filings on AWS
The Form 990 e-filed data were first made available in 2016 via the [Amazon Web Services Public Dataset ](https://docs.opendata.aws/irs-990/readme.html) program as individual representations of each e-file in XML format. These millions of files each contain hundreds of details about a nonprofit’s spending, revenue, compensation, fundraising and more.  

* Sources：https://docs.opendata.aws/irs-990/readme.html
* Codebooks :  this file is a index of 990 fillings on AWS, no codebooks. 
* Data formats: string

```
# read AWS data
import pandas as pd
index_2011 = pd.read_csv("https://s3.amazonaws.com/irs-form-990/index_2011.csv")

#sample of index_2011
sample_index_2011 = index_2011.sample(n=10000, replace=True)
print(sample_index_2011.head(5))
```

## Database 3: National Center for Charitable Statistics Data (NCCS) 

NCCS derives its data files primarily from information that tax-exempt nonprofit organizations
file with the Internal Revenue Service. IRS databases offer the most comprehensive standardized data on tax-exempt organizations but have significant limitations. The data produced by NCCS are intended for use by researchers and policy-makers in their quantitative analyses, and as a springboard for more in-depth survey or case study research.

* Sources: https://nccs-data.urban.org/index.php
* Codebooks: https://nccs-data.urban.org/dd2.php?close=1&form=Core+2013+PC+Full+990
* Data formats: string




```
# read NCCS data coreco.core2015co_full990.csv

import pandas as pd
NCCS_2015 = pd.read_csv("https://nccs-data.urban.org/data/core/2015/coreco.core2015co_full990.csv")
len(NCCS_2015)

# sample of NCCS_2015
sample_NCCS_2015 = NCCS_2015.sample(n=10000, replace=True)
print(sample_NCCS_2015.head(5))
len(sample_NCCS_2015)
list(sample_NCCS_2015)
```

# Merge  Dadabase 2 and Database 3

```
#Merge index_2011 and NCCS_2015
AWS_NCCS_2015 = index_2011.merge(NCCS_2015, on="EIN", how="inner")
print(AWS_NCCS_2015.head(5))
len(AWS_NCCS_2015)
list(AWS_NCCS_2015)
```

```
#Merge sample_index_2011 and sample_NCCS_2015
sample_AWS_NCCS_2015 = sample_index_2011.merge(sample_NCCS_2015, on="EIN", how="inner")
print(sample_AWS_NCCS_2015.head(5))
list(sample_AWS_NCCS_2015)
len(sample_AWS_NCCS_2015)
```
